#### Import appropriate packages. 

```python
%matplotlib inline 
import numpy as np 
import sep 
from astropy.io import fits 
import matplotlib.pyplot as plt
```

#### Assign files to variable 

```python
f105w_file= 'https://archive.stsci.edu/pub/hlsp/hudf12/hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits'
f125w_file= 'https://archive.stsci.edu/pub/hlsp/hudf12/hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits'
f160w_file= 'https://archive.stsci.edu/pub/hlsp/hudf12/hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits'
```

#### Open the files and save the data to a variable.

```python
f105w= fits.open(f105w_file)[0].data
f125w= fits.open(f125w_file)[0].data
f160w= fits.open(f160w_file)[0].data
```

#### Convert the HSV image to RGB.

```python
import matplotlib.colors as colors
rgb_image = np.zeros((3600,3600,3))

rgb_image[:,:,0] = f105w
rgb_image[:,:,1] = f125w
rgb_image[:,:,2] = f160w

rgb_image = colors.hsv_to_rgb(rgb_image)
```

#### Plot the image.

```python
f = plt.figure(figsize=(7,7))
plt.imshow(rgb_image)
plt.savefig("RGB_image.png")
```
